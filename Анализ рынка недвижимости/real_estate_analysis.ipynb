### Пропуски значений

Сначала посчитаем, сколько в таблице пропущенных значений. Для этого достаточно двух методов pandas: isna() и sum().

```python
# подсчёт пропусков в таблице: 
df.isna().sum()
```

Пропуски данных обнаружены в четырнадцати колонках.

У нас изначално по каждой квартире на продажу доступны два вида данных:

- city_centers_nearest - расстояние до центра города (м)
- parks_around_3000 - число парков в радиусе 3 км

В связи с чем причины появления пропусков в данных тоже разные.

Так пропуски в данных вписанных пользователем могли появиться потому, что у человека отсутствуют указанные данные, например нет балкона, нет данных о высоте потолков, либо просто значения были пропущены при заполнении данных, например не указано назавание населенного пункта. Однако такие пропуски могут появиться в результате сбоя при выгрузке данных из базы.

Пропуски в данных полученных автоматически на основе картографических данных могли появиться в связи с тем, что произошел сбой при выгрузке данных из базы, либо произошел сбой в автоматизированной системе, либо эти данные отсутсвуют в автоматизированной системе.
___

**Замена пропусков в колонке `balcony`  - число балконов.**
Для некоторых пропущенных значений можно предположить логичную замену.
Например, если человек не указал число балконов — скорее всего, их нет. Такие пропуски правильно заменить на 0 методом fillna().

```python
# замена пропусков в balcony
df['balcony'] = df['balcony'].fillna(0)

# проверка, сосчитаем строки без значения балконов
df['balcony'].isna().sum()
```

**Замена пропусков в колонке `is_apartment` - апартаменты.**
Не все пропущенные значения влияют на исследование.
Так как апартаментов менее 2% от общего числа объектов недвижимости, этими данными можно было пренебречь, они не важны для нашей работы. 
Однако для пропущенных значений в колонке 'is_apartment' тоже можно предположить логичную замену. Если человек пропустил это значение, скорее всего это не апартаменты. Такие пропуски правильно заменить на 'False':

```python
# замена пропусков в is_apartment
df['is_apartment'] = df['is_apartment'].fillna(False)

# проверка, сосчитаем пропущенные строки в колонке апартаментов 
df['is_apartment'].isna().sum()
```

**Замена пропусков в колонках `floors_total` - всего этажей в доме и `locality_name` - название населённого пункта.** 
Для этих типов данных нет подходящего значения на замену. В этом случае правильно оставить эти значения пустыми, заменить их явными обозначениями или удалить.

В столбце `'floors_total'` всего 86 пропусков - это менее 1% от объема даннных, заполнять их нолями нет смысла, можно удалить, т.к. их мало, то в общем виде данные не исказятся.

Пропущенных значений в столбце `'locality_name'` 49, что также менее 1%, лучше их удалить. 

Сначала посморим на распереление пропусков в таблице, если пропуски распределены равномерно, то удаляем их, сделаем допущение, что на результаты это не повлияет.

```python
# проверим распределение пропусков в floors_total
df[df['floors_total'].isna()].head(50)
```

Этот код похож на фильтр. Выражение `df['floors_total'].isna()` проверяет все строки в столбце floors_total и оставляет только те, в которых есть NaN. Прпуски в floors_total распределены равномерно, можно удалить методом dropna() и восстановим индексы методом reset_index() c параметром drop, равным True. 

```python
# удаление пропусков в floors_total  и locality_name
df = df.dropna(subset=['floors_total', 'locality_name']).reset_index(drop=True) 
```

Высота потолков, жилая площадь, площадь кухни - количественные переменные. Пропуски в таких переменных заполняют характерными значениями. Это значения, характеризующие состояние выборки, — набора данных, выбранных для проведения исследования. Чтобы примерно оценить типичные значения выборки, годятся среднее арифметическое mean() или медиана median().

**Замена пропусков в колонке `ceiling_height` - выстота потолков.**
В колонке с высотой потолков обнаружено 9195 пропусков - это почти 39% от общего объема данных в таблице. Хоть эти данные не повлияют на настоящее иследование, просто удалить пропущенные значения нельзя, мы потеряем большой объем данных. 

Многоквартирирные дома строятся в основном по типовым проектам, а выстота потолков в квартирах реагламентируется законом (СНиПами и ГОСТами). В среднем, этот параметр колеблется в пределах 2.4–3.2 метра. Минимальная высота потолка определяется СНиП как 2.5 м. Такие потолки характерны для хрущевок. В новостройках их минимальная высота составляет 2.6 метра. А в элитных домах она может быть более 3 метров.

## Расчёты и добавление результатов в таблицу

**Расчет цены квадратного метра:**
Для расчета цены квадратного метра разделим значения столбца `'last_price'` на значения столбца `'total_area'` и добавим полученные значения в новый столбец датафрейма `'price_meter'` сразу изменив тип данных нового столбца с `float64` на `int64`.

```python
# расчет стоимости квадратного метра и добавление нового столбца в датафрейм
df['price_meter'] = (df['last_price'] / df['total_area']).astype('int').round(0)
```

**Расчет дня недели, месяца и года публикации объявления:**
Чтобы осуществить такой расчёт, нужно поместить время в класс DatetimeIndex и применить к нему атрибут month, dayofweek, year:

```python
# расчет дня недели и добавление в новый столбец dayofweek
df['dayofweek'] = pd.DatetimeIndex(df['first_day_exposition']).dayofweek

# расчет месяца и добавление в новый столбец month
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month

# расчет года и добавление в новый столбец year сокращенным методом
df['year'] = df['first_day_exposition'].dt.year
```

**Расчет этажа квартиры:**
Требуется расчитать этаж квартиры; варианты — первый, последний, другой.


```python
# создаем функцию
def floor(floor_number):
     if floor_number['floor'] == floor_number['floors_total']:
         return 'последний'
     if floor_number['floor']  == 1:
         return 'первый'
     return 'другой'
    
# создаем в таблице df новый столбец 'floors_sort' и добавим в него данные функции     
df['floors_sort'] = df.apply(floor, axis=1)
```

**Соотношение жилой и общей площади, а также отношение площади кухни к общей.**
Для каждой квартиры посчитаем долю жилой площади как отношение жилой к общей площади квартиры. Для этого разделим значения столбца `'living_area'` на `total_area` и округлим получившееся значение методом `round()` с 1 знаком после запятой. Затем добавим данные в новый столбец датафрейма `'living_total_ratio'`.

Также посчитаем долю кухни в общей площади квартиры. Для этого разделим значения столбца `'kitchen_area'` на `total_area`.

```python
# отношение жилой к общей площади и добавление нового столбца в датафрейм
df['living_total_ratio'] = ((df['living_area'] / df['total_area']) * 100).round(1)

# отношение площади кухни к общей площади и добавление нового столбца в датафрейм
df['kitchen_total_ratio'] = ((df['kitchen_area'] / df['total_area']) * 100).round(1)

# проверка добавления столбцов в датафрейм
display(df.head(5)) 
```

Изучим совместное распределение величин. Взаимосвязь двух или нескольких величин называется «корреляция». График (да и жизненный опыт) позволяет утверждать, что цена и площадь взаимосвязаны, или коррелируют. Оценим численно, насколько тесна взаимосвязь. Для этого находят коэффициент корреляции Пирсона. Он помогает определить наличие линейной связи между величинами и принимает значения от -1 до 1.
- Если с ростом первой величины растёт вторая, то коэффициент корреляции Пирсона — положительный.
- Если при изменении одной величины другая остаётся прежней, то коэффициент равен 0.
- Если рост одной величины связан с уменьшением другой, коэффициент отрицательный.


```python
# найдем коэффициент корреляции цены и площади
df['total_area'].corr(df['last_price'])
```

Корреляция в 0.65 говорит о наличии связи. Выходит, увеличение площади сопровождается прибавкой в стоимости, но не всегда. Обратное же утверждение не имеет смысла, т.к. при увеличении стоимости площадь квартиры увелитится.

```python
# посторим график зависимости цены и числа комнат
df.plot(x='rooms', y='last_price', kind='scatter')

plt.title('График зависимости цены от числа комнат'+ "\n")
plt.show()
```

```python
# найдем коэффициент корреляции цены и числа комнат
df['last_price'].corr(df['rooms'])
```

Корреляция 0.45 приближен положительна, значит, линейная связь есть, но слабая. Выходит, что увеличение числа комнат сопровождается увеличением стоимости квартиры, но не всегда. Бывает, что коэффициент нулевой не потому, что связи между значениями нет, а из-за того, что у неё более сложный, не линейный характер. Потому-то коэффициент корреляции такую связь не берёт.
Однако на графике наблюдается снижение стоимости с увеличением числа комнат. Самыми дорогими являются 3-4 комнатные квартиры до одного миллиона за квадратный метер, однако встречается стоимость однакомнатных квартир также приближается к этой отметке.

```python
# создадим столбец с расстоянием до центра в километрах: округлим до целых значений
df['city_centers_km'] = np.ceil(df['city_centers_nearest'] / 1000)

# посчитаем среднюю цену для каждого километра
# построим график

(
    df.query('locality_name == "Санкт-Петербург"')
    .pivot_table(index=['city_centers_km'], values='last_price')
    .reset_index()
    .plot(x='city_centers_km', y='last_price', style='o-')
)

plt.title('График зависимости цены от удалённости от центра'+ "\n")
plt.show()
```

По графику определеяется центральная зона в районе 5-8 км.
